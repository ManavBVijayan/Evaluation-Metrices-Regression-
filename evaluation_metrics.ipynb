```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate example data
np.random.seed(42)
actual_values = np.random.rand(100) * 10  # Actual values (ground truth)
predicted_values = actual_values + np.random.randn(100)  # Predicted values with some noise
```

```python
# Mean Absolute Error (MAE)
mae = np.mean(np.abs(actual_values - predicted_values))

# Mean Squared Error (MSE)
mse = np.mean((actual_values - predicted_values)**2)

# R-squared
ss_total = np.sum((actual_values - np.mean(actual_values))**2)
ss_residual = np.sum((actual_values - predicted_values)**2)
r_squared = 1 - (ss_residual / ss_total)
```

```python
# Print the results
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared: {r_squared:.4f}")
```

```python
# Scatter plot to visualize actual vs predicted values
plt.scatter(actual_values, predicted_values, color='blue', label='Actual vs Predicted')
plt.plot([0, 10], [0, 10], color='red', linestyle='--', linewidth=2, label='Ideal Line')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Regression Evaluation')
plt.legend()
plt.show()
```
